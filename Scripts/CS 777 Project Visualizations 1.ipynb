```python
import sys

import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
## Visualising the distance travelled yearly by flights

df_1 = pd.read_csv('/content/drive/MyDrive/777_project/Distance_vs_Year.csv')
x = df_1['Year']
y = df_1['Distance']
plt.figure(figsize=(7, 5), dpi=200)
plt.title('Yearly Distance travelled')
plt.xlabel('Year')
plt.ylabel('Distance Travelled in Miles')
plt.bar(x,y, edgecolor='black')
plt.ticklabel_format(style='plain')
```

## We can see that the highest number of distance travelled by flights was in 2019 and lowest was in 2022

```python
## Number of flights every month

df_2 = pd.read_csv('/content/drive/MyDrive/777_project/Number_flights_vs_Months.csv')
x = df_2['Month']
y = df_2['Number of Flights']
plt.figure(figsize=(8, 5), dpi=200)
plt.title('Number of Flights Monthly')
plt.xlabel('Months')
plt.ylabel('Number of Flights')
plt.bar(x,y, edgecolor='black')
plt.ticklabel_format(style='plain')
plt.ylim((0,3000000))
plt.xticks(ticks=list(range(1,13)))
```

## We can see that January and March are among the most travelled months.

```python
## Number of flights daily

df_3 = pd.read_csv('/content/drive/MyDrive/777_project/Number_flights_vs_Weeks.csv')
x = df_3['Week']
y = df_3['Number of Flights']
plt.figure(figsize=(8, 5), dpi=200)
plt.title('Number of Flights Daily')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Flights')
plt.bar(x,y, edgecolor='black')
plt.ticklabel_format(style='plain')
```

```python
## Visualising delay groups.


df_4 = pd.read_csv('/content/drive/MyDrive/777_project/Number_flights_vs_DelayGroup.csv')
plt.figure(figsize=(8,6))
df_4.sort_values('Number of Flights', inplace=True)
x = df_4['DelayGroup']
y = df_4['Number of Flights']
plt.figure(figsize=(8, 5), dpi=200)
plt.title('Delay Groups')
plt.xlabel('DelayGroups')
plt.ylabel('Number of Flights')
plt.bar(x,y, edgecolor='black')
```
